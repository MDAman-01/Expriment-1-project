```python
import random
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = {
    "pH": [round(random.uniform(6.5, 8.5), 2) for _ in range(300)],
    "turbidity": [round(random.uniform(1, 10), 2) for _ in range(300)],
    "DO": [round(random.uniform(6, 9), 2) for _ in range(300)],
    "BOD": [round(random.uniform(3, 5), 2) for _ in range(300)],
    "conductivity": [round(random.uniform(100, 1000), 2) for _ in range(300)],
}

# Water Quality Rule (1 = Safe, 0 = Unsafe)
def quality(pH, turb, DO, BOD, cond):
    if (6.5 <= pH <= 8.5) and turb < 5 and DO > 6 and BOD < 3 and cond < 400:
        return 1   # Safe
    return 0       # Unsafe

# Add quality column
data["water_quality"] = [
    quality(data["pH"][i], data["turbidity"][i], data["DO"][i], data["BOD"][i], data["conductivity"][i])
    for i in range(300)
]

# Convert to DataFrame
df = pd.DataFrame(data)
```

```python
count_data = df["water_quality"].value_counts()
plt.bar(count_data.index, count_data.values)
plt.title("Safe vs Unsafe Water Samples")
plt.xlabel("Quality (1=Safe, 0=Unsafe)")
plt.ylabel("Count")
```

```python
data = [df["pH"], df["DO"], df["BOD"], df["turbidity"]]
colors = ['green', 'red', 'yellow', 'blue']  # Different color for each box

plt.figure(figsize=(12,8))
bp = plt.boxplot(data, patch_artist=True)  # patch_artist=True allows coloring

# Apply different colors to each box
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(df["pH"], label="pH", color="red")
plt.plot(df["turbidity"], label="Turbidity", color="blue")
plt.plot(df["DO"], label="Dissolved Oxygen", color="green")
plt.plot(df["BOD"], label="BOD", color="purple")

plt.title("Line Graph Comparison of All Water Parameters")
plt.xlabel("Sample Number")
plt.ylabel("Parameter Value")
plt.legend()
plt.grid(True)

plt.show()
```
